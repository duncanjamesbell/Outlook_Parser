```python
import win32com.client
import win32com
import os
import sys
import pandas as pd
import datetime
```

```python
outlook = win32com.client.Dispatch('outlook.application')
mapi = outlook.GetNamespace("MAPI")
for account in mapi.Accounts:
    print(account.DeliveryStore.DisplayName)
```

```python
for s in mapi.Stores:
    if 'Resources Assn Trends' in s.DisplayName:
        inbox = s.GetDefaultFolder(6)
```

```python
inbox.items
```

```python
Subject = []
SenderEmailAddress = []
SentOnBehalfOfName = []
SentOn = []
AutoForwarded = []
Body = []

keywords = ['retire', 
            'no longer', 
            'new email', 
            'has left', 
            'have left', 
            'last day', 
            'passed away', 
            'died', 
            'email address',
           'email account']

emails = pd.DataFrame()
for m in inbox.Items:
    try:
        sent = m.SentOn.date()
        if sent >= datetime.date(2021,3,1):
            if m.SentOn.date() <= datetime.date(2021,3,31):
                if any(x in str(m.Body).lower() for x in keywords):
                    try:
                        #test that each attribute is available before appending to list
                        m.Subject
                        m.SenderEmailAddress
                        m.SentOnBehalfOfName
                        m.SentOn
                        m.AutoForwarded
                        m.Body
                        Subject.append(m.Subject)
                        SenderEmailAddress.append(m.SenderEmailAddress)
                        SentOnBehalfOfName.append(m.SentOnBehalfOfName)
                        SentOn.append(m.SentOn)
                        AutoForwarded.append(m.AutoForwarded)
                        Body.append(m.Body)
                    except:
                        pass
        else:
            break
    except:
        pass
    
emails['Subject'] = Subject
emails['SenderEmailAddress'] = SenderEmailAddress
emails['SentOnBehalfOfName'] = SentOnBehalfOfName
emails['SentOn'] = SentOn
emails['AutoForwarded'] = AutoForwarded
emails['Body'] = Body
```

```python
emails.to_csv('ATresources 3.2021.csv')
```
